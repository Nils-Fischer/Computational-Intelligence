```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)

# Aufgabe 7.2
def f(x: float):
    return np.cos(x / 3) + np.sin(5 / (np.abs(x) + 0.2)) - 0.3 * x

def g(x: float, a: list):
    return sum([a[p] * x**p for p in range(0, 10)])

x = np.linspace(-5, 5, 1001)
a_0 = np.array([-1/2, 1/4, -1/8, 1/16, -1/32, 1/64, -1/128, 1/256, -1/512, 1/1024])

error = lambda a: np.sum(np.abs(f(x) - g(x, a))) / 1001
print(f"durchschnittlicher Fehler: {error(a_0)}")

# visualisieren
plt.plot(x, f(x), label="f(x)")
plt.plot(x, g(x, a_0), label="g(x)")
plt.xlabel("x")
plt.xlabel("y")
plt.title("f(x) und g(x)")
plt.ylim(-10, 10)
plt.legend()
plt.show()
```

```python
min_fehler = float("inf")
min_a = []
for _ in range(1001):
    a = np.random.uniform(-5, 5, 10)
    fehler = error(a)
    if fehler < min_fehler:
        min_fehler = fehler
        min_a = a
print(f"beste gefundene Koeffizienten mit Fehlerwert: {min_fehler}, sind: \n{min_a}")
```

```python
min_a = []
min_fehler = float("inf")
for _ in range(11):
    a = np.random.uniform(-1, 1, 10)
    fehler = error(a)
    for _ in range(101):
        a_new = a + np.random.uniform(-0.1, 0.1, 10)
        fehler_new = error(a_new)
        if fehler > fehler_new:
            fehler = fehler_new
            a = a_new
    if min_fehler > fehler:
        min_fehler = fehler
        min_a = a
print(f"beste gefundene Koeffizienten mit Fehlerwert: {min_fehler}, sind: \n{min_a}")
plt.plot(x, g(x, min_a), label="g(x)")
plt.plot(x, f(x), label="f(x)")
plt.ylim(-10,10)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("g(x) mit gefundenem a und f(x)")
plt.show()

```
